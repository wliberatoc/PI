```
import matplotlib.pyplot as plt  
import numpy as np              

x = np.arange(0,1*np.pi-1,0.1)   
seno = np.sin(x) #seno
coss = np.cos(x) #cosseno

plt.plot(x,seno,x,coss)
plt.ylabel('sin(x) and cos(x)')
plt.title('Plot of sin and cos from 0 to 4pi')
plt.legend(['seno(x)', 'cosseno(x)'])      
plt.show()
```
